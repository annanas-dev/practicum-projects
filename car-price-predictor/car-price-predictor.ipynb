## Исследовательский анализ данных

```python
# Функция для распределения признаков по типу данных
def distribution_by_data_types(df):
    
    # Категориальные признаки 
    categorical_columns_names = df.select_dtypes(include='object').columns.tolist()
    # Количественные признаки
    quantitative_columns_names = df.select_dtypes(include='number').columns.tolist()
    # Даты
    date_columns_names = df.select_dtypes(include='datetime64').columns.tolist()
    
    return categorical_columns_names, quantitative_columns_names, date_columns_names
```

```python
# Функция для анализа категориального признака
def research_categorical_sign(df, categorical_column):
    
    colors = sns.color_palette('mako')
    plt.figure(figsize=(8,8))
    
    t = df[categorical_column].value_counts().reset_index().rename(columns={'index': categorical_column, categorical_column: 'count'})
    
    if len(df[categorical_column].unique().tolist()) <= 5:
        # Построение круговой диаграммы, если меньше 6 категорий
        plt.pie(t['count'], labels=t[categorical_column], autopct='%1.1f%%', colors=colors, textprops={'fontsize': 14} )
        plt.title(f'Круговая диаграмма для признака {categorical_column}',  fontsize=18)
        plt.show()
        print(t)
    else:
        # Построение столбчатой диаграммы, если категорий больше 6
        more_9 = df.copy()
        frequent_categories = t.iloc[:30][categorical_column].tolist()
        more_9.loc[~more_9[categorical_column].isin(frequent_categories), categorical_column] = 'other categories'
        category_order = more_9[categorical_column].value_counts().index.tolist()
        plt.title(f'Столбчатая диаграмма для столбца \'{categorical_column}\'', fontsize=16)
        sns.countplot(data=more_9, x=categorical_column, order=category_order, palette='mako')
        plt.xticks(rotation=90, ha='center')
        plt.show()
        print(more_9[categorical_column].value_counts(normalize=True))
```

```python
# Функция для анализа непрерывного количественного признака
def research_quantitative_sign(df, quantitative_column):
    
    if quantitative_column in ['registration_month', 'kilometer']: # Данные столбцы рациональней рассмотреть как дискретную количественную величину
        
        # Построение столбчатой диаграммы 
        fig, axs = plt.subplots(2, 2, figsize=(18, 6))
        
        plt.title(f'Столбчатая диаграмма для столбца \'{quantitative_column}\'', fontsize=16)
        sns.countplot(data=df, x=quantitative_column, palette='mako', ax=axs[0][0])
        
    else:
        
        # Построение гистограммы
        fig, axs = plt.subplots(2, 2, figsize=(18, 6)) 
    
        sns.histplot(x=quantitative_column, data=df, ax=axs[0][0], palette='mako', bins=20)
        axs[0][0].set_ylabel('Количество')
        axs[0][0].set_title(f'Гистограмма для признака {quantitative_column}', fontsize=16)
    
        # Построение ящика с усами
        sns.boxplot(x=quantitative_column, data=df, ax=axs[0][1], palette='mako')
        axs[0][1].set_title('Ящик с усами', fontsize=16)
    
    # Применение CDF для проверки вида распределения (и для того, чтобы убедиться с какой величиной мы работаем: дискретной или непрервыной)
    sns.ecdfplot(data=df[quantitative_column], ax=axs[1][0])
    axs[1][0].set_title('CDF')
        
    # QQ-plot для проверки нормальности распределения
    stats.probplot(df[quantitative_column], dist='norm', plot=plt)
    axs[1][1].set_title('QQ-plot')

    fig.suptitle(f'Распределение для столбца \'{quantitative_column}\'', fontsize=18)
    
    plt.show()

    # Описание данных
    print(f'Описание данных для столбца \'{quantitative_column}\':\n{df[quantitative_column].describe()}')
```

```python
# Функция для анализа дат
def research_data_signs(df, discrete_columns):
    
    fig, axs = plt.subplots(1, 3, figsize=(24, 6))  
    
    # Построение столбчатой диаграммы
    
    # День
    plt.title(f'Диаграммы для столбца \'{discrete_columns}\' ДЕНЬ', fontsize=16)
    sns.countplot(data=df, x=pd.DatetimeIndex(df[discrete_columns]).day, palette='mako', ax=axs[0])
    axs[0].set_ylabel('Количество')
    axs[0].set_title('Дни')
    
    # Месяц
    plt.title(f'Диаграммы для столбца \'{discrete_columns}\' МЕСЯЦ', fontsize=16)
    sns.countplot(data=df, x=pd.DatetimeIndex(df[discrete_columns]).month, palette='mako', ax=axs[1])
    axs[1].set_ylabel('Количество')
    axs[1].set_title('Месяцы')
    
    # Год
    plt.title(f'Диаграммы для столбца \'{discrete_columns}\' ГОД', fontsize=16)
    sns.countplot(data=df, x=pd.DatetimeIndex(df[discrete_columns]).year, palette='mako', ax=axs[2])
    axs[2].set_ylabel('Количество')
    axs[2].set_title('Года')
    
    fig.suptitle(f'Диаграммы для столбца \'{discrete_columns}\'', fontsize=18)
    
    plt.show() 
```

```python
# Функция для проведения исследовательского анализа
def research_analysis(df):
    
    cat_col_names, num_col_names, date_col_names = distribution_by_data_types(df)
    
    for num_col in num_col_names:
        research_quantitative_sign(df, num_col)
    
    for cat_col in cat_col_names:
        research_categorical_sign(df, cat_col)
        
    for date_col in date_col_names:
        research_data_signs(df, date_col)
```

```python
research_analysis(autos.drop('postal_code', axis=1))
```

## Корреляционный анализ данных

```python
# Функция для проведения корреляционного анализа
def correlation_analysis(df, inter_col):
    
    # Рассчитаем корреляционную матрицу
    phik_overview = phik_matrix(df, interval_cols=inter_col)
    
    plt.figure(figsize=(20, 15))
    
    # Визуализируем корреляционную матрицу
    sns.heatmap(phik_overview, annot=True, fmt=".4f", cmap='BuPu', vmin=0, vmax=1, 
                cbar_kws={'label': r'correlation $\phi_K$'})
    
    plt.title(r'correlation $\phi_K$', fontsize=18)
    plt.show()
```

В датафреме есть несколько столбцов в формате datetime64, который препятствует корреляционному анализу. Преобразуем даты в Unix Time (количество секунд с 1 января 1970 года).

```python
# Преобразуем даты в Unix Time (количество секунд с 1 января 1970 года)
autos_corr = autos.copy()
autos_corr['date_crawled_unix'] = autos_corr['date_crawled'].astype(int) // 10**9
autos_corr['date_created_unix'] = autos_corr['date_created'].astype(int) // 10**9
autos_corr['last_seen_unix'] = autos_corr['last_seen'].astype(int) // 10**9

autos_corr = autos_corr.drop('date_crawled', axis=1)
autos_corr = autos_corr.drop('date_created', axis=1)
autos_corr = autos_corr.drop('last_seen', axis=1)

autos_corr.head()
```

## Обучение моделей

```python
# Создадим список, в который будем добавлять все обученные модели для последующего сравнения
trained_models = []
```

### Линейная регрессия

```python
cat_col_names, num_col_names, date_col_names = distribution_by_data_types(X)

display(cat_col_names,
        num_col_names)
```

```python
# Разделим категориальные признаки в зависимости от способа кодирования
ohe_columns = cat_col_names
num_columns = num_col_names
```

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
)
```

```python
# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
])
```

```python
param_grid_LinearRegression = [
    # словарь для модели LinearRegression()
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']    
    }
]
```

```python
randomized_search_LinearRegression = RandomizedSearchCV(
    pipe_final, 
    param_grid_LinearRegression, 
    cv=5,
    n_iter=4,
    scoring='neg_root_mean_squared_error',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
warnings.filterwarnings('ignore')

randomized_search_LinearRegression.fit(X_train, y_train)
```

```python
# Добавим модель в список
trained_models.append(randomized_search_LinearRegression)
```

### Дерево решений

```python
param_grid_DecisionTreeRegressor = [
    # словарь для модели DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 20),
        'models__max_features': range(2, 20),
        'models__min_samples_split': range(2, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']    
    }
]
```

```python
randomized_search_DecisionTreeRegressor = RandomizedSearchCV(
    pipe_final, 
    param_grid_DecisionTreeRegressor, 
    cv=5,
    scoring='neg_root_mean_squared_error',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
warnings.filterwarnings('ignore')

randomized_search_DecisionTreeRegressor.fit(X_train, y_train)
```

```python
trained_models.append(randomized_search_DecisionTreeRegressor)
```

### LightGBM

```python
# создаём пайплайн для подготовки признаков с помощью Ordinal-кодирование
# SimpleImputer + OE
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
# создаём общий пайплайн для подготовки данных
data_preprocessor_for_lgbm = ColumnTransformer(
    [
     ('ord', ord_pipe, ohe_columns)
    ], 
    remainder='passthrough'
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final_for_lgbm = Pipeline([
    ('preprocessor_for_lgbm', data_preprocessor_for_lgbm),
    ('models', LinearRegression())
])
```

```python
param_grid_LGBMRegressor = [
    # словарь для модели LGBMRegressor()
    {
        'models': [LGBMRegressor(random_state=RANDOM_STATE, objective='regression', metric='rmse')],
        'models__max_depth': [1, 5],  
        'models__learning_rate': [0.01, 0.1, 1.0],
    }
]
```

```python
randomized_search_LGBMRegressor = RandomizedSearchCV(
    pipe_final_for_lgbm, 
    param_distributions=param_grid_LGBMRegressor,  
    cv=5,
    scoring='neg_root_mean_squared_error',  
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
randomized_search_LGBMRegressor.fit(X_train, y_train)
```

```python
trained_models.append(randomized_search_LGBMRegressor)
```

### CatBoost

```python
model_catboost = CatBoostRegressor(cat_features=ohe_columns,
                                   random_state=RANDOM_STATE,
                                   verbose=100)
```

```python
param_grid_CatBoostRegressor = [
    # словарь для модели CatBoostRegressor()
    {
        'depth': [1, 5],
        'learning_rate': [0.1, 1.0]
    }
]
```

```python
randomized_search_CatBoostRegressor = RandomizedSearchCV(
    estimator=model_catboost, 
    param_distributions=param_grid_CatBoostRegressor,  
    cv=5,
    scoring='neg_root_mean_squared_error',  
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
randomized_search_CatBoostRegressor.fit(X_train, y_train)
```

## Анализ моделей

```python
RMSE = []
training_time = []
prediction_time = []
```

```python
for model in trained_models:
    RMSE.append(-model.best_score_)
    best_model = model.best_estimator_
    start_of_training = time.time()
    best_model.fit(X_train, y_train)
    finish_of_training = time.time()
    training_time.append(finish_of_training - start_of_training)
    start_of_prediction = time.time()
    best_model.predict(X_test)
    finish_of_prediction = time.time()
    prediction_time.append(finish_of_prediction - start_of_prediction)
```

```python
model_analysis = pd.DataFrame({'Модель': ['LinearRegression', 'DecisionTreeRegressor', 'LightGBM', 'CatBoostRegressor'],
                               'RMSE': RMSE,
                               'Время обучения': training_time,
                               'Время предсказания': prediction_time})
model_analysis
```

- **Связи входных признаков**
  - Мощность автомобиля (power), тип коробки передач (gearbox) и тип автомобильного кузова (vehicle_type) имеют умеренную линейную зависимость от марки автомобиля (brand).
  - Умеренная линейная зависимость также наблюдается между признаками gearbox и моделью автомобиля (model).
  - Еще одна средняя линейная зависимость заметна между годом регистрации (registration_year) и типом кузова (vehicle_type), а также  между годом регистрации (registration_year) и моделью автомобиля (model).
  
  - Сильная линейная зависимость отмечается между датой последней активности пользователя (last_seen_unix) и датой скачивания анкеты (date_crawled_unix).
  - Признак power имеет значительную зависимость от модели автомобиля (model).
  
  - Мультиколлинеарность четко выражена между model и brand: коэффициент корреляции равен 1, что указывает на линейную зависимость — каждая модель соответствует своему бренду. По признакам model и brand было принято решение объединить их в единый признак model_brand.
  - Также сильная корреляция (коэффициент 0.91) наблюдается между признаком vehicle_type и model, что также указывает на мультиколлинеарность. Удалять не стоит, так как у нас могут быть две абсолютно одинаковых модели, отличающиеся ровно на этот признак и сильно отличающиеся при этом по цене (солярис, хетчбек и седан). Для 80% автомобилей у нас, допустим, нет вариантов разных 'vehicle_type', для остальных 20% признак будет существенным.

  - Разница между качеством предсказания DecisionTreeRegressor и LightGBM незначительна - модель занимает 2 место по качеству предсказаний среди всех вариантов моделей, как и по времени обучения, однако демонстрирует самое продолжительное время предсказания, однако разница между LightGBM, LinearRegression и CatBoost.
- **LinearRegression**:
  - Модель по всем метрикам занимает предпоследнюю позицию - по качеству предсказания хуже, чем LightGBM и CatBoost, по продолжительности времени предсказания и обучения лучше самой плохой модели, но хуже прочих.


LightGBM обеспечивает достойное качество предсказаний, что является главным критерием. Время обучения у модели второе по сккорости, но предсказания выполняются медленнее других. Лучше нее по первым двум метрикам - CatBoost, ео разница не велика, а также CatBoost очень сильно проигрывает по времени обучения. Таким образом, LightGBM - компромисс между точностью и временем работы.
