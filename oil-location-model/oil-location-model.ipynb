# Выбор локации для скважины

## Описание проекта

Допустим, вы работаете в добывающей компании «ГлавРосГосНефть». Нужно решить, где бурить новую скважину.

Вам предоставлены пробы нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов. Постройте модель машинного обучения, которая поможет определить регион, где добыча принесёт наибольшую прибыль. Проанализируйте возможную прибыль и риски техникой *Bootstrap.*

Шаги для выбора локации:

- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.

## Описание данных

Данные геологоразведки трёх регионов находятся в файлах: 
- `datasets/geo_data_0.csv`
- `datasets/geo_data_1.csv`
- `datasets/geo_data_2.csv`

Признаки, содержащиеся в датасетах:
- `id` — уникальный идентификатор скважины;
- `f0, f1, f2` — три признака точек (неважно, что они означают, но сами признаки значимы);
- `product` — объём запасов в скважине (тыс. баррелей).

**Условия задачи:**
- При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
- Бюджет на разработку скважин в регионе — 10 млрд рублей.
- При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.
- После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.
- Данные синтетические: детали контрактов и характеристики месторождений не разглашаются.

```python
# Функция для изучения датафрейма и получения общей информации о нем
def studying_dataframe(df):
    display(df.info(),
           df.head())
```

```python
# Автоматизация проверки датафрейма на наличе дубликатов
# На входе функции подается комбинация столбцов, по которой можно проверить наличие НЕПОЛНЫХ явных дубликатов
def search_for_duplicates(df, column_combinations):
    
    # Проверка на наличие ПОЛНЫХ явных дубликатов
    print(f'Количество ПОЛНЫХ явных дубликатов: {df.duplicated().sum()}\n')
    
    # Проверка на наличие НЕПОЛНЫХ явных дубликатов
    if len(column_combinations) > 0:
        count_of_dup = df[column_combinations].duplicated().sum()
        print(f'Количество НЕПОЛНЫХ явных дубликатов: {count_of_dup}\n')
    
    # Проверка на наличие неявных дубликаов
    m = 'Проверка на неявные дубликаты:'
    if len(df.select_dtypes(exclude='number').columns.tolist()) > 0:
        for cat_col in df.select_dtypes(exclude='number').columns.tolist():
            if cat_col != 'id':
                print(m)
                print(f'- Для {cat_col}: {df[cat_col].unique().tolist()}')
                m = 0
```

```python
for name_df, df in studied_dataframes.items():
    print('*' * 120)
    print(f'Изучение датафрейма "{name_df}":\n')
    studying_dataframe(df)
```

В каждом из рассматриваемых датафреймов: 
- пропуски отсутствуют, о чем мы можем судить по соответсвию количества Non-Null значений в каждом из признаков (столбцов) общему количеству строк
- во всех столбцах типы данных соответствуют содержащимся в них значениям
- названия столбцов соответсвуют единому змеиному стилю

## Исследовательский анализ данных

Проведем исследовательский анализ данных для каждого из трех регионов.

```python
# Функция для исследования непрерывных количественных признаков
def research_quantitative_continuous_signs(df):
    
    for col in df.columns.tolist():
        fig, axs = plt.subplots(2, 2, figsize=(18, 11))
        
        # Построение гистограммы
        sns.histplot(data=df, x=col, ax=axs[0][0], palette='viridis', bins=50)
        axs[0][0].set_ylabel('Количество')
        axs[0][0].set_title('Гистограмма')
        
        # Построение ящика с усами
        sns.boxplot(data=df, x=col, ax=axs[0][1], palette='viridis')
        axs[0][1].set_title('Ящик с усами')
        
        # Применение CDF для проверки вида распределения
        sns.ecdfplot(data=df[col], ax=axs[1][0])
        axs[1][0].set_title('CDF')
        
        # QQ-plot для проверки нормальности распределения
        stats.probplot(df[col], dist='norm', plot=plt)
        axs[1][1].set_title('QQ-plot')
        
        # Заголовок для всего рисунка
        fig.suptitle(f'Распределение для столбца \'{col}\'', fontsize=16)
        
        # Показать графики
        plt.show()
        
        # Описание данных
        print(f'Описание данных для столбца \'{col}\':\n{df[col].describe()}')
```

## Корреляционный анализ данных

```python
# Функция для проведения корреляционного анализа
def correlation_analysis(df):

    # Рассчитаем корреляционную матрицу
    phik_overview = phik_matrix(df, interval_cols=df.columns.tolist())
    
    # Визуализируем корреляционную матрицу
    plot_correlation_matrix(phik_overview.values,
                        x_labels=phik_overview.columns,
                        y_labels=phik_overview.index,
                        vmin=0, vmax=1, color_map='BuPu',
                        title=r'correlation $\phi_K$',
                        fontsize_factor=1.5,
                        figsize=(10, 8))
    
    # Матрица диаграмм рассеяния
    sns.pairplot(df, palette='viridis', diag_kind='kde', markers='o')
    plt.show()
```

## Обучение и проверка модели

```python
def prediction_linear_regression(df):
    
    X = df.drop('product', axis=1)
    y = df['product']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=RANDOM_STATE)
    
    # Кодирование данных
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # Обучение модели
    model_lr = LinearRegression()
    model_lr.fit(X_train, y_train)
    predictions = model_lr.predict(X_test)
    
    # Расчет среднего запаса предсказанного сырья
    mean_prediction_resource_reserve = predictions.mean()
    
    # Расчет RMSE
    rmse_score = root_mean_squared_error(y_test, predictions)
    
    return y_test, predictions, mean_prediction_resource_reserve, rmse_score
```

```python
# Регион 0
y_valid_0, y_pred_0, mean_pred_resource_0, rmse_0 = prediction_linear_regression(geo_data_0)
print(f'{"*" * 120}\nРегион 0:\nСредний запас предсказанного сырья = {mean_pred_resource_0}\nRMSE = {rmse_0}')

# Регион 1
y_valid_1, y_pred_1, mean_pred_resource_1, rmse_1 = prediction_linear_regression(geo_data_1)
print(f'{"*" * 120}\nРегион 1:\nСредний запас предсказанного сырья = {mean_pred_resource_1}\nRMSE = {rmse_1}')

# Регион 2
y_valid_2, y_pred_2, mean_pred_resource_2, rmse_2 = prediction_linear_regression(geo_data_2)
print(f'{"*" * 120}\nРегион 2:\nСредний запас предсказанного сырья = {mean_pred_resource_2}\nRMSE = {rmse_2}')
```

Анализ результатов работы модели показал следующее:
- средний объем запасов сырья в разных регионах варьируется значительно. В регионе 2 наблюдается самый высокий средний объем запасов, который составляет 94.93 тыс. баррелей по предсказанным данным. В регионе 0 этот показатель немного ниже и составляет 92.6 тыс. баррелей. Наименьший средний объем запасов зарегистрирован в регионе 1, где он равен 68.58 тыс. баррелей. Эти различия могут быть связаны с вариациями геологических характеристик в каждом регионе.
- качество модели также различается в зависимости от региона. Модель, обученная на данных георазведки региона 1, демонстрирует наивысшую точность (относительно других моделей). Это совпадает с результатами корреляционного анализа, который показал наибольшее количество ярко выраженных зависимостей между целевой переменной и входными признаками именно в этом регионе. Что касается двух других моделей - в контексте данной задачи RMSE имеют приемлимые значения.

## Подготовка к расчёту прибыли

```python
WELLS = 500 # при разведке региона исследуют 500 точек
BEST_WELLS = 200 # выбирают 200 лучших скважин для разработки
BUDGET = 10000000000 # бюджет
PRICE_PER_UNIT = 450000 # цена за единицу продукта
```

Рассчитаем достаточный объём сырья для безубыточной разработки новой скважины. 

```python
BUDGET / PRICE_PER_UNIT # количество единиц сырья необходимого для безубыточной разработки в целом
```

```python
sufficient_volume = (BUDGET / PRICE_PER_UNIT) / BEST_WELLS
print(f'Объём сырья для безубыточной разработки новой скважины = {sufficient_volume:.2f} тыс. баррелей')
```

Сравним полученный объём сырья со средним запасом в каждом регионе.

```python
average_resource_reserve_by_region = {'0': mean_pred_resource_0,
                                     '1': mean_pred_resource_1,
                                     '2': mean_pred_resource_2}

for num_region, mean_pred_resource in average_resource_reserve_by_region.items():
    difference = 1 - mean_pred_resource / sufficient_volume
    if difference < 0:
        print(f'В регионе {num_region} средний запас сырья БОЛЬШЕ достаточного объёма сырья для безубыточной разработки новой скважины на {abs(difference):%}\n')
    else:
        print(f'В регионе {num_region} средний запас сырья МЕНЬШЕ достаточного объёма сырья для безубыточной разработки новой скважины на {abs(difference):%}\n')
        
```

Во всех рассматриваемых регионах средний запас сырья НЕ превышает достаточный объём сырья для безубыточной разработки новой скважины, однако разница не так велика.

```python
target_pred_by_region = {'0': (y_valid_0, y_pred_0),
                         '1': (y_valid_1, y_pred_1),
                         '2': (y_valid_2, y_pred_2)}

for num_region, (y_valid, y_pred) in target_pred_by_region.items():
    print(f'Количество скважин в регионе {num_region}, в которых запас сырья превышает достаточный объем = {y_valid[y_valid > sufficient_volume].count()}\n')
```

В каждом из рассматриваемых регионов набирается достаточное количество скважин (более 200), запасы сырья в которых превышают достаточный объём сырья для безубыточной разработки новой скважины. 

## Расчёт прибыли и рисков 

Рассчитаем прибыль по выбранным скважинам и предсказаниям модели.

```python
# Функция для расчета прибыли 
def profit_calculation(real_volume, prediction_volume):
    prediction_volume = prediction_volume.sort_values(ascending=False)
    profit = real_volume[prediction_volume.index][:BEST_WELLS].sum() * PRICE_PER_UNIT - BUDGET
    return profit
```

```python
for num_region, (y_valid, y_pred) in target_pred_by_region.items():
    print(f'В регионе {num_region} прибыль составляет {profit_calculation(y_valid, pd.Series(y_pred, index=y_valid.index))}\n')
```

Согласно нашим подсчетам, наибольшую прибыль принесет разработка новых скважин в регионе 0. 

Посчитаем риски и прибыль для каждого региона:
1. Применим технику Bootstrap с 1000 выборок, чтобы найти распределение прибыли.
2. Найдем среднюю прибыль, 95%-й доверительный интервал и риск убытков.

```python
def bootstrap_profit(target, prognosis):
    state = np.random.RandomState(12345)
    values=[]
    
    for i in range(1000):
        subsample = target.sample(n=WELLS, replace=True, random_state=state)
        prognosis_after_bootstrap = prognosis[subsample.index]
        
        values.append(profit_calculation(subsample, prognosis_after_bootstrap))
    
    values = pd.Series(values)
    
    fig, axs = plt.subplots(figsize=(8, 5))
    
    # Построим гистограмму распределения прибыли
    sns.histplot(values, ax=axs, palette='viridis', bins=50)
    axs.set_title('Гистограмма распределения прибыли по региону')
    axs.set_xlabel('Прибыль')
    axs.set_ylabel('Количество')
    
    # Найдем среднюю прибыль
    mean_profit = values.mean()
    axs.axvline(x=mean_profit, color='red', linestyle='--', linewidth=5)
    
    # Найдем 95%-й доверительный интервал
    lower = values.quantile(0.025)
    upper = values.quantile(0.975) 
    
    axs.axvline(x=lower, color='blue', linestyle='--', linewidth=3)
    axs.axvline(x=upper, color='blue', linestyle='--', linewidth=3)
    
    # Найдем риск убытков
    risk = np.mean(values < 0)
    
    plt.show()
    
    results = {
        'Значение': [mean_profit, f'({lower:.2f}, {upper:.2f})', f'{risk:.2%}']
    }
    
    index = ['Средняя прибыль', 'Доверительный интервал', 'Риск']
    
    results_df = pd.DataFrame(results, index=index)
    print(results_df)
```

```python
for num_region, (y_valid, y_pred) in target_pred_by_region.items():
    print('*' * 120)
    print(f'\nРегион {num_region}')
    bootstrap_profit(y_valid, pd.Series(y_pred, index=y_valid.index))
```

Применение техники технику Bootstrap и поиск средней прибыли, 95%-й доверительного интервала и риск убытков дают нам следующую информацию:
- Средняя прибыль
  - Регион 1 демонстрирует наибольшую среднюю прибыль, что делает его самым прибыльным регионом.
  - Регион 0 следует за регионом 1, показывая немного меньшую прибыль.
  - Регион 2 имеет наименьший показатель средней прибыли среди всех регионов.
  

  
- Риск убатков
  - Регион 1 является наименее рискованным в плане убытков, с риском, составляющим всего 0.7%.
  - Регион 0 имеет риск убытков в размере 5.3%.
  - Регион 2 показывает наибольший риск убытков, равный 7.8%.
